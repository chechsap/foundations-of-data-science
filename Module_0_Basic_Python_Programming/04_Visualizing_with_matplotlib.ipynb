## 1. Plotting


### 1.1 Line plot

Say we want to plot a sequential data, we can plot using the following code snippet:

```python
import numpy as np

np.random.seed(42)

x = np.arange(10)
y = np.random.randint(50, 100, size=10)

plt.figure(figsize=(6, 4), dpi=100)
plt.ylabel('value')
plt.xlabel('time')
plt.title('Line Plot')

plt.plot(x, y)
plt.show()
```

### 1.2 Scatter plot

We can also do a scatter plot for the data above.

```python
np.random.seed(42)

x = np.arange(10)
y = np.random.randint(50, 100, size=10)

plt.figure(figsize=(6, 4), dpi=100)
plt.ylabel('value')
plt.xlabel('time')
plt.title('Scatter Plot')

plt.scatter(x, y)
plt.show()
```

What if we have multiple categories in our data? In the following example, there are 3 categories.

```python
from sklearn import datasets

# import some data to play with
iris = datasets.load_iris()
X = iris.data[:, :2]  # we only take the first two features.
y = iris.target

plt.figure(figsize=(6, 4), dpi=100)
plt.ylabel('sepal width')
plt.xlabel('sepal length')
plt.title('Scatter Plot')

plt.scatter(X[:,0], X[:,1], c=y, cmap=plt.cm.Set1)
plt.show()
```

### 1.3 Bar graph

Bar graph is used for categorical data.

```python
x = ['Jude', 'Sash', 'Unisse']
y = [5, 6, 4]

plt.figure(figsize=(6, 4), dpi=100)
plt.ylabel('height')
plt.xlabel('person')
plt.title('Height')

plt.bar(x, y)
plt.show()
```

### 1.4 Histogram

Histograms are typically used for showing the distribution of data.

```python
dist = np.random.normal(size=100)

plt.figure(figsize=(6, 4), dpi=100)
plt.hist(dist)
plt.show()
```
